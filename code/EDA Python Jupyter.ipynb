```python
import pandas as pd
from azure.storage.blob import BlobServiceClient
import io

connection_string = ""
container_name = "raw-organized"

# Nombre del blob dentro del container
blob_path = "customers/ingestion_date=2025-12-06/olist_customers_dataset.csv"

# Crear cliente del servicio Blob
blob_service_client = BlobServiceClient.from_connection_string(connection_string)
print("Conexión OK", blob_service_client)
container_client = blob_service_client.get_container_client(container=container_name)

# Print blob names 
blob_list = container_client.list_blobs()
for blob in blob_list: 
   print(blob.name)


# 1. Downloading single CSV -------------------------------------- 

# 2. Elige uno manualmente (copiado de la lista anterior)
blob_name = "customers/ingestion_date=2025-12-06/olist_customers_dataset.csv "

# 3. Obtener blob client
blob_client = container_client.get_blob_client(blob_name)
print("Blob client generated for: ", blob_client.blob_name)

# Leer Blob 
stream = blob_client.download_blob().readall()
print("Stream concluded")

# Crear DataFrame del CSV
df = pd.read_csv(io.BytesIO(stream))
df.head()


# 2. Downloading all CSVs ------------------------ 
df_list = {}
for blob in blob_list: 
   blob_name = blob.name
   blob_client = container_client.get_blob_client(blob_name)
   print("Blob client generated for: ", blob_client.blob_name)

   # Leer Blob 
   stream = blob_client.download_blob().readall()
   print("Stream concluded for: ", blob_client.blob_name)

   # Crear DataFrame del CSV
   df = pd.read_csv(io.BytesIO(stream))
   
   # Añadirla a la lista de daraframes 
   df_list[blob_name] = df


   df.head()

print(df_list)

```
